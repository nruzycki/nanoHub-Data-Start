```python
# Load the Python librariers you will need here we will load Pandas (data tables), Numpy (math),SciPy (stats), Matplotlib (plotting) 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#Here we will set up a data frame for use in Python and use Pandas to make the frame
#setting up the data frame for height We are using a random generator here to create 
data = pd.DataFrame()
data['gender'] = ['male'] * 100 + ['female'] * 100
data['height'] = np.append(np.random.normal(69, 8, 100), np.random.normal(64, 5, 100))
data['weight'] = np.append(np.random.normal(195, 25, 100), np.random.normal(166, 15, 100))
data
```

#### not very useful, but plotted

```python
#You can use a term called "groupby" to try and regroup your table (note the table stays resorted)
data.groupby('gender')
# describe the data
data.groupby('gender').describe()
```

```python
# to box plot your data, you can use the following code
data.groupby('gender').boxplot(column=None, ax=None, fontsize=None, rot=0, grid=True, figsize=None, layout=None, return_type=None)
plt.savefig('mf.png', bbox_inches='tight')
```

## T-Tests in Python

```python
## First I will create a subset for males and females from the data set. I am using this https://pythonfordatascience.org/independent-t-test-python/ as a reference
#create a male and female data table
female = data[(data['gender'] == 'female')]
female.reset_index(inplace= True)

male = data[(data['gender'] == 'male')]
male.reset_index(inplace= True)
```

```python
# run a t test on the data
stats.ttest_ind(male['height'], female['height'])
```
